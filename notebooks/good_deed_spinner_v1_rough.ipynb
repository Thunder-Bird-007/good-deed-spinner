# Good Deed Spinner

This project is a beginner-friendly Python app inspired by random spin games, but redesigned in a meaningful and halal way.

Instead of betting money, the program:
- generates 3 random symbols
- checks whether they match
- gives the user a useful task
- awards points

```python
import random

SYMBOLS = ["🌙", "📘", "🤲", "⭐", "🕌"]
TASKS = [
    "Make dua for your parents.",
    "Read 5 lines of Qur'an.",
    "Say Astaghfirullah 33 times.",
    "Revise one important formula.",
    "Drink a glass of water.",
    "Help a family member with something small.",
    "Study with focus for 10 minutes.",
    "Clean a small part of your desk."
]   

def spin_symbols():
    return [
        random.choice(SYMBOLS),
        random.choice(SYMBOLS),
        random.choice(SYMBOLS)
    ]

def get_random_task():
    return random.choice(TASKS)

def get_result(spin_result):
    a, b, c = spin_result
    if a==b==c:
        return 20, "Jackpot! You got three of the same symbol!"
    elif a==b or b==c or c==a:
        return 10, "Two of the same symbol! Not bad!"
    else:
        return 5, "No matches this time. But you still get a good deed task!"

def run_spin():
    spin_result = spin_symbols()
    score, msg = get_result(spin_result)
    task = get_random_task()

    print("Spinning...")
    print(" | ".join(spin_result))
    print(msg)
    print(f"Your score: {score}")
    print(f"Your good deed task: {task}")

    return score

total_score = 0
for round in range(3):
    print(f"\n--- Spin {round + 1} ---")
    total_score += run_spin()
print(f"\nTotal score: {total_score}")
```
